```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#PART A. 

cauchy_example = np.random.standard_cauchy(10000)
plt.hist(cauchy_example, bins=100, range=(-10,10))
plt.show()
```

```python
#PART B

normal_distribution = np.random.standard_normal(10000)
plt.hist(normal_distribution, bins=100, range=(-10,10))
plt.show()
```

```python
#PART C

partial_sum=[]
for i in range(1,10001):
    x= sum(cauchy_example[0:i])/i
    partial_sum.append(x)
    
plt.plot(np.arange(1,10001), partial_sum)
plt.show()

#Notice that as many times as you repeat the code, the value of the graph tends to tread toward 0
#This makes intuitive sense because a Cauchy Distrubution is heavy tailed and does not have an Expected Value
```

```python
#For Normal Distribution 

partial_sum=[]
for i in range(1,10001):
    x= sum(normal_distribution[0:i])/i
    partial_sum.append(x)
    
plt.plot(np.arange(1,10001), partial_sum)
plt.show()
```
