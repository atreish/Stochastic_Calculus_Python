```python
import numpy as np
from matplotlib import pyplot as plt



for i in range(10):
    print('path %s' % i)
    increments = np.random.normal(0, 1/10, 100)
    dB = np.append([0], increments)
    B = np.cumsum(dB)
    ts = np.arange(0, 101) / 100
    X = np.exp(ts / 2) * np.cos(B)
    y = np.exp(ts / 2) * np.sin(B)
    y = np.append([0], y)[:-1]
    Y = 1 - np.cumsum(y * dB)
    plt.plot(ts, X)
    plt.show()
    plt.plot(ts, Y)
    plt.show()
```

```python
plt.plot(ts, X)
plt.plot(ts, Y)
plt.legend(['X', 'Y'])
```

```python
for l in [10, 100, 1000]:
    avg = []
    for i in range(10):
        increments = np.random.normal(0, 1/10, 100)
        dB = np.append([0], increments)
        B = np.cumsum(dB)
        ts = np.arange(0, 101) / 100
        X = np.exp(ts / 2) * np.cos(B)
        y = np.exp(ts / 2) * np.sin(B)
        Y = 1 - np.cumsum(y * dB)
        avg.append(np.abs(X[-1] - Y[-1]))
    avg = np.array(avg).mean()
    print('%d paths\' estimate: %s' % (l, avg))
    
```
