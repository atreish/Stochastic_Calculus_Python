```python
import numpy as	np;
import numpy.random as nr;
import scipy.linalg as LA;
import matplotlib.pyplot as plt;
import math;
```

```python
def fbm_covariance(s: float, t: float, h: float):
    return (1/2)*((t**(2*h)) + (s**(2*h)) -(abs(t-s)**(2*h)))
```

```python
timeincrements = np.arange(0.01, 1.01, 0.01) #create a range of values for i and j
```

```python
#Fractional Brownian Motion with H=.01

fbm1_cov = np.reshape(np.array([fbm_covariance(i,j, h=.01) for i in timeincrements for j in timeincrements]), (100,100))
```

```python
fbm1_plot = np.array([np.linalg.cholesky(fbm1_cov).dot(np.random.normal(loc=0, scale=1, size=100)) for _ in range(100)])
```

```python
plt.figure(figsize=(15,8))
plt.title('Fractional Brownian Motion with Hurst Index = 0.01')
for i in fbm1_plot:
    plt.plot(i)
```
