```python
import numpy as np
from matplotlib import pyplot as plt

def f(x, y, h):
    s = min(x, y)
    t = max(x, y)
    return 0.5*(t**(2 * h) + s **(2 * h) - np.abs(t - s) ** (2 * h))


for h in [0.1, 0.5, 0.9]:
    cov_mat = np.zeros([100, 100])
    for i in range(100):
        for j in range(100):
            cov_mat[i, j] = f(i + 1, j + 1, h)
    cov_mat = cov_mat / 100
    for i in range(100):
        Z = np.random.normal(0, 1, 100)
        A = np.linalg.cholesky(cov_mat)
        X = np.append([0], A@Z)
        plt.plot(np.linspace(0, 1, 101), X)
    plt.show()
```
