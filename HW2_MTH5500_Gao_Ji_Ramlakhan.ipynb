```python
def brownian_c(s, t):
  return min(s,t)
```

```python
def brownian_b(s, t):
  if s<t:
    return s*(1-t)
  else: 
    return t*(1-s)
```

```python
def ornstein_uhlenbeck(s, t):
  if s<t:
    return 0.5*math.exp(-2*(t-s))*(1-math.exp(-2*s))
  else: 
    return 0.5*math.exp(-2*(s-t))*(1-math.exp(-2*t))
```

```python
def ornstein_uhlenbeck_s(s, t):
  if s<t:
    return 0.5*math.exp(-2*(t-s))
  else: 
    return 0.5*math.exp(-2*(s-t))
```

```python
bm_cov = np.tril(np.ones([100, 100])) @ np.triu(np.ones([100, 100])) / 100
```

```python
bm_plot = np.array([np.linalg.cholesky(bm_cov).dot(np.random.normal(loc=0, scale=1, size=100)) for _ in range (100)])
```

```python
plt.figure(figsize=(10, 6))
plt.title('Standard Brownian Motion')
for i in bm_plot: 
  plt.plot(i)
```

```python
bb_cov = np.zeros([100, 100])
for i in range(100):
  for j in range(100):
    s = min(i, j) + 1
    t = max(i, j) + 1
    bb_cov[i, j] = s * (100 - t)
bb_cov /= 100 
bb_cov += 0.01
```

```python
sample = [np.random.normal(loc=0, scale=1, size=100) for _ in range(100)]

bb_plot = np.array([np.linalg.cholesky(bb_cov).dot(sample) for _ in sample])
plt.figure(figsize=(10, 6))
plt.title('Brownian Bridge')
for i in bb_plot: 
  plt.plot(i)
```

```python
ou_cov = np.zeros([100, 100])
for i in range(1, 101):
  for j in range(1, 101):
    ou_cov[i - 1, j - 1] = ornstein_uhlenbeck(i / 100, j / 100)
```

```python
ou_plot = np.array([np.linalg.cholesky(bb_cov).dot(np.random.normal(loc=0, scale=1, size=100)) for _ in range (100)])
plt.figure(figsize=(10, 6))
plt.title('Ornstein Uhlenbeck')
for i in ou_plot: 
  plt.plot(i)
```

```python
ou_cov = np.zeros([100, 100])
for i in range(1, 101):
  for j in range(1, 101):
    ou_cov[i - 1, j - 1] = ornstein_uhlenbeck_s(i / 100, j / 100)

ou_plot = np.array([np.linalg.cholesky(bb_cov).dot(np.random.normal(loc=0, scale=1, size=100)) for _ in range (100)])
plt.figure(figsize=(10, 6))
plt.title('Ornstein Uhlenbeck Variance 1/2')
for i in ou_plot: 
  plt.plot(i)
```
